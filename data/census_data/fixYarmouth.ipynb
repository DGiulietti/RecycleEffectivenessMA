```python
import pandas as pd
```

```python
groups = {
    'education_char':'DP02',
    'economic_char':'DP03',
    'housing_char':'DP04',
    'demo_char':'DP05'
}

for characteristic in groups:
    exec(f'{characteristic} = pd.read_csv("{characteristic}.csv", index_col="municipality")')
    
place_char = pd.concat([education_char,economic_char,housing_char,demo_char], axis=1)
```

```python
yar_char = place_char.loc[['West Yarmouth', 'Yarmouth Port', 'South Yarmouth']]
```

```python
# df[["A", "B"]].multiply(df["C"], axis="index")

yar_char[['edu_high_school_and_higher_%', 'edu_bachelors_and_higher_%']] = yar_char[['edu_high_school_and_higher_%', 'edu_bachelors_and_higher_%']].multiply(yar_char['KEY_population_25_and_older'], axis='index')
```

```python
yar_char[['unemployment_rate_%', 'median_household_income',
       'mean_household_income', 'families_below_poverty_level_%',
       'people_below_poverty_level_%']] = yar_char[['unemployment_rate_%', 'median_household_income',
       'mean_household_income', 'families_below_poverty_level_%',
       'people_below_poverty_level_%']].multiply(yar_char['KEY_pop_16_+'], axis='index')
```

```python
yar_char[['occupancy_%', '1-unit_detached_%',
       '1-unit_attached_%', '2-units_%', '3-4_units_%', '5-9_units_%',
       '10-19_units_%', '20+_units_%', 'mobile_home_%', 'boat_RV_van_%',
       'owner-occupied_%', 'renter-occupied_%']] = yar_char[['occupancy_%', '1-unit_detached_%',
       '1-unit_attached_%', '2-units_%', '3-4_units_%', '5-9_units_%',
       '10-19_units_%', '20+_units_%', 'mobile_home_%', 'boat_RV_van_%',
       'owner-occupied_%', 'renter-occupied_%']].multiply(yar_char['KEY_total_housing_units'], axis='index')
```

```python
yar_char[['males-to-100females', 'median_age',
       'demo_white_%', 'demo_black_aa_%', 'demo_american_indian_%',
       'demo_asian_%', 'demo_native_islander%', 'demo_other_Race_%',
       'demo_hispanic_latino_%']] = yar_char[['males-to-100females', 'median_age',
       'demo_white_%', 'demo_black_aa_%', 'demo_american_indian_%',
       'demo_asian_%', 'demo_native_islander%', 'demo_other_Race_%',
       'demo_hispanic_latino_%']].multiply(yar_char['KEY_total_population'], axis='index')
```

```python
import pandas as pd
```

```python
vote_char = pd.read_csv('data/MA_voting_info/2019voter_data.csv', usecols=['municipality','dem%','rep%','un%'])
vote_char.head()
```

```python
vote_char['municipality'] = vote_char['municipality'].str.replace('.','')
vote_char.head()
```

```python
vote_char['municipality'] = vote_char['municipality'].str.lower()
vote_char.head()
```

```python
vote_char['municipality'] = vote_char['municipality'].str.title()
vote_char.head()
```

```python
vote_char.set_index('municipality', inplace = True)
vote_char.head()
```
